### 실습19. 커스텀 파이프라인 구현

```python
import torch
from torch.nn.functional import softmax
from transformers import AutoModelForSequenceClassification, AutoTokenizer

class CustomPipeline:
    def __init__(self, model_id):
        self.model = AutoModelForSequenceClassification.from_pretrained(model_id)#목적에 맞는 모델 헤드 불러옴
        self.tokenizer = AutoTokenizer.from_pretrained(model_id)#모델과 동일한 토크나이저 불러옴
        self.model.eval() #모델 평가 모드(아직은 몰라도 됨.)

    def __call__(self, texts):
        tokenized = self.tokenizer(texts, return_tensors="pt", padding=True, truncation=True)# 입력 텍스트를 토크나이저를 사용하여 토큰화
        # return_tensors="pt": PyTorch 텐서로 반환, padding=True: 패딩 추가, truncation=True: 잘라내기

        with torch.no_grad():#기울기 계산 비활성화(옵티마이저 계산 생략)-> 추론 과정에서는 옵티마이징 할 필요 없음.
            outputs = self.model(**tokenized)#**연산자는 딕셔너리를 키워드 인자로 unpacking하여 모델에 입력. 즉 토큰 아이디
            logits = outputs.logits# 모델 출력에서 logits 값을 추출. (분류되지 않은 예측값)

        #추론이 정확할 확률 구하기
        probabilities = softmax(logits, dim=-1)
        scores, labels = torch.max(probabilities, dim=-1)
        labels_str = [self.model.config.id2label[label_idx] for label_idx in labels.tolist()]

        return [{"label": label, "score": score.item()} for label, score in zip(labels_str, scores)]

custom_pipeline = CustomPipeline(model_id)
print(custom_pipeline(test_dataset['title'][:5]))
print(test_dataset['label_str'][:5])
```

```python
import torch
from torch.nn.functional import softmax
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from datasets import ClassLabel

# ------------------------------------------------------
# 1) 데이터셋에 사람이 읽을 라벨 문자열(label_str) 컬럼을 보장
#    - 우선순위: 데이터셋 ClassLabel → 모델 config.id2label
# ------------------------------------------------------
def ensure_label_str_column(dataset, model):
    if "label_str" in dataset.column_names:
        return dataset  # 이미 있으면 그대로 사용

    if "label" not in dataset.column_names:
        raise ValueError("데이터셋에 'label' 컬럼이 없습니다. label_str를 만들 수 없어요.")

    # (1) 데이터셋의 라벨 스키마(ClassLabel)가 있으면 그것을 사용 (가장 안전)
    features = dataset.features
    if isinstance(features.get("label"), ClassLabel):
        int2str = features["label"].int2str
        def _add_label_str(ex):
            return {"label_str": int2str(ex["label"])}
        return dataset.map(_add_label_str)

    # (2) 없으면 모델의 id2label 사용 (주의: 데이터셋 라벨 인덱스와 모델 라벨 인덱스 불일치 가능)
    #     키가 문자열일 수 있으므로 int 키로 강제 변환
    id2label = {int(k): v for k, v in model.config.id2label.items()}
    def _add_label_str(ex):
        return {"label_str": id2label[int(ex["label"])]}
    return dataset.map(_add_label_str)

# ------------------------------------------------------
# 2) 커스텀 파이프라인
#    - 단일 문자열/리스트 입력 모두 지원
#    - 디바이스 자동 선택
# ------------------------------------------------------
class CustomPipeline:
    def __init__(self, model_id):
        self.model = AutoModelForSequenceClassification.from_pretrained(model_id)
        self.tokenizer = AutoTokenizer.from_pretrained(model_id)
        self.model.eval()

        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.model.to(self.device)

        # id2label 키가 문자열일 수 있으므로 미리 정규화
        self.id2label = {int(k): v for k, v in self.model.config.id2label.items()}

    def __call__(self, texts):
        # 단일 문자열도 리스트로 변환
        if isinstance(texts, str):
            texts = [texts]

        tokenized = self.tokenizer(
            texts,
            return_tensors="pt",
            padding=True,
            truncation=True
        )
        tokenized = {k: v.to(self.device) for k, v in tokenized.items()}

        with torch.no_grad():
            outputs = self.model(**tokenized)
            logits = outputs.logits

        probabilities = softmax(logits, dim=-1)            # [batch, num_labels]
        scores, labels = torch.max(probabilities, dim=-1)  # argmax

        # 라벨 문자열로 변환
        labels_str = [self.id2label[int(i)] for i in labels.tolist()]

        return [{"label": lbl, "score": float(sc)} for lbl, sc in zip(labels_str, scores.tolist())]

# ------------------------------------------------------
# 3) 실행 예시
# ------------------------------------------------------
# model_id = "distilbert-base-uncased-finetuned-sst-2-english"  # 예: 감성분류 모델
custom_pipeline = CustomPipeline(model_id)

# test_dataset에 label_str 보장
test_dataset = ensure_label_str_column(test_dataset, custom_pipeline.model)

# 예측
preds = custom_pipeline(test_dataset["title"][:5])
golds = test_dataset["label_str"][:5]

# 보기 좋게 출력
for i, (p, g, t) in enumerate(zip(preds, golds, test_dataset["title"][:5])):
    print(f"[{i}] title: {t}")
    print(f"    pred : {p['label']} ({p['score']:.3f})")
    print(f"    gold : {g}")
```
